### Problem 1:

1. Write a program to calculate area of shapes. Your program should be capable of calculating the area of a square, rectangle, triangle and a circle. The user should be presented with options to select the shape. Based on which shape is chosed by the user, the program should ask for the appropriate input and print the resulting area on the screen.

When the program is run, the screen should display something like this:

**Which shape would you like to calculate the area for? Please enter the option number-**<br>
**1. Square**<br>
**2. Rectangle**<br>
**3. Triangle**<br>
**4. Circle**<br>
**Enter Option: _**

Say the user enters the option 1.

**Please enter the length of a side: _**

If the user enters a value of 5. The output should be:

**The area of the square is 25**

This program should indicate that the input is invalid if the user enters a character instead of a number as input. For instance if the user enters a value of ‘a’ instead of 5 in the previous example the program should prompt:

**Invalid input, please enter a number: _**


```python
def square_area(a):
    area = a**2
    return area

def rectangle_area(l,b): 
    area = l*b
    return area

def triangle_area(b,h):
    area = (1/2)*b*h
    return area
    
def circle_area(r):
    import math 
    area = math.pi*(r**2)
    return area
```

```python
print('Which shape would you like to calculate the area for? Please enter the option number:')
print('1. Square')
print('2. Rectangle')
print('3. Triangle')
print('4. Circle')

for i in range(4):
    uinput = input('Enter your option : ')
    print()

    if uinput.isdigit():
        option = int(uinput)

        if option == 1:
        
            print('SQUARE')
        
            length = float(input('Please enter the length of a side : '))
        
            result = square_area(length)
                      
            print('The area of the square is {:.2f} square units'.format(result))
            print()
    
        elif option == 2:
        
            print('RECTANGLE')
    
            length = float(input('Please enter the length : '))
            breadth = float(input('Please enter the breadth : '))
        
            result = rectangle_area(length,breadth)
        
            print('The area of the rectangle is {:.2f} square units'.format(result))
            print()
    
        elif option == 3:
        
            print('TRIANGLE')
    
            breadth = float(input('Please enter the breadth : '))
            height = float(input('Please enter the height: '))
        
            result = triangle_area(breadth,height)
        
            print('The area of the triangle is {:.2f} square units'.format(result))
            print()
    
        elif option == 4:
        
            print('CIRCLE')
    
            radius = float(input('Please enter the radius : '))
        
            result = circle_area(radius)
        
            print('The area of the circle is {:.2f} square units'.format(result)) 
            print()
       
        else:
            print('Invalid input, please enter a number between 1 and 4 : ') 
            print()
    else:
        print('Invalid input, please enter a number : ')
        print()
    
```

### Problem 2:

Create and encrypter in python based on the ceaser cipher. It is a substitution cipher where each character of the original text is shifted a certain number characters in the alphabet. Write a function that would require 2 arguments – the input text to be encrypted and a key. For eg: Given the input text ‘hello’ and the key 3, the resulting encryted text would be ‘khoor’. Here you can see that every character in the string hello is shifted by 3 characters. ‘h’ has shifted to ‘k’, ‘e’ has shifted to ‘h’ and so on. If a key of 5 were used, the resulting string would be ‘mjqqt’. This function should be capable of ignoring any characters which are not alphabets. Th2 character ‘z’ entered b
y the user for a key of 3 would result in ‘c’.

Usage:<br>
**encrypt(‘hello world!’, 3)**<br>
**‘khoor zruog!’**

Similarly create decrypter which can decode the encryted text when provided the input text and key

Usage:<br>
**decrypt(‘khoor zruog!’, 3)**<br>
**‘hello world!’**

For the sake of simplicity you can assume that input solely consists of lowercase alphabets, spaces and punctuation symbols. Numbers in the input text would also be ignored similar to symbols.

```python
def encrypt(text,key):
    
    encoded_char = ''
    
    for i in range(len(text)):
        character = text[i]
                
        if(character.islower()):
                    
            position = ord(character)  # Finding the position between 0-25
            
            actual_index = position - ord('a') 
               
            new_index = (actual_index + key) % 26 # Shifting
          
            new_code = new_index + ord('a')  # Fetching the new code after shifting
            
            new_char = chr(new_code) # Converting new code to character
            
            encoded_char = encoded_char + new_char    # Append to encrypted string 
            
        elif(character.isupper()):
            
            position = ord(character) 
            
            actual_index = position - ord('A')
           
            new_index = (actual_index + key) % 26 
          
            new_code = new_index + ord('A')  
            
            new_char = chr(new_code) 
            
            encoded_char = encoded_char + new_char
            
        else:
            
            encoded_char = encoded_char + character  # All other characters other than alphabets will be displayed as it is.
                  
    return encoded_char

def decrypt(encrypted_text, key):
    
    decrypted_text = ''
    
    for i in range(len(encrypted_text)):
        
        character = encrypted_text[i]
        
        if (character.islower()):
            
            position = ord(character)

            actual_index = position - ord("a")

            new_index = (actual_index - key) % 26 #reverse shifting

            new_code = new_index + ord("a")

            new_char = chr(new_code)

            decrypted_text = decrypted_text + new_char
            
        elif(character.isupper()):
            
            position = ord(character)

            actual_index = position - ord("A")

            new_index = (actual_index - key) % 26

            new_code = new_index + ord("A")

            new_char = chr(new_code)

            decrypted_text = decrypted_text + new_char
            
        else:
            
            decrypted_text = decrypted_text + character
            
    return decrypted_text
    
```

```python
print('1. Encryption')
print('2. Decryption')
print()

for i in range(2):
    option = int(input('Enter your option : '))
    print()

    if option == 1:
        text = input('Enter the text to be encrypted: ')
        key = int(input('Enter the no: of positions to be shifted: '))
        print()
        result = encrypt(text,key)
    
        print(f'Encrypted text for {text} ----- {result}')
        print()
          
    else:
        
        encrypted_text = input('Enter the text to be decrypted: ')
        key = int(input('Enter the no: of positions to be decrypted: '))
        print()
        result = decrypt(encrypted_text,key)
          
        print(f'Decrypted text for {encrypted_text} ----- {result}')
        print()
          
```

```python
Note : Using the modulo operation ensures that the encryption formula evaluates to a number that never exceeds the upper bound of 25.
```
